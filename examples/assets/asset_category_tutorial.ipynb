# Asset Category API Tutorial

This notebook creates an asset universe through the public `msm.api.assets` row classes with example auto-registration enabled.

```python
import os
import sys
from pathlib import Path

PROJECT_ROOT = next(
    (path for path in [Path.cwd(), *Path.cwd().parents] if (path / "src" / "msm").exists()),
    Path.cwd(),
)
sys.path[:0] = [str(PROJECT_ROOT / "src"), str(PROJECT_ROOT)]

from examples.platform.bootstrap import (  # noqa: E402
    EXAMPLE_AUTO_REGISTER_ENV,
    EXAMPLE_METATABLE_NAMESPACE,
)

os.environ.setdefault(EXAMPLE_AUTO_REGISTER_ENV, EXAMPLE_METATABLE_NAMESPACE)

from msm.api.assets import Asset, AssetCategory  # noqa: E402
```

```python
btc = Asset.upsert(unique_identifier="example-category-btc", asset_type="crypto")
eth = Asset.upsert(unique_identifier="example-category-eth", asset_type="crypto")

category = AssetCategory.upsert(
    unique_identifier="example-category-crypto-majors",
    display_name="Example Crypto Majors",
    description="Example-scoped crypto asset universe.",
)
memberships = AssetCategory.replace_memberships(
    category_uid=category.uid,
    asset_uids=[btc.uid, eth.uid],
)

{
    "category": category,
    "memberships": memberships,
}
```
